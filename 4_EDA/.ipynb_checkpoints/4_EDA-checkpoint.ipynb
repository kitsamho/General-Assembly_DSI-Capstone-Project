## `Helper Functions`
---

Some helper functions - as we are running a few visualisations by each brand (remember we have seven!) then these will keep the notebook tidier

```python
#returns a frame with any given metric and a custom range for that metric
#e.g.Dataframe containing only Asda posts that have between 0 and 20 comments
def frame_mask(brand, metric,metric_max=np.inf,metric_min=0):
    subset = df[(df.Brand == brand) & (df[metric] <= metric_max) & (df[metric] >=metric_min)]
    return subset[metric]

#returns a subset of the dataframe filtered on any brand(s) needed
#e.g. Dataframe containing only Asda and Waitrose rows
def brand_mask(brand,df=df):
    try:
        filter_list = [each.lower() for each in brand.split(',')]
    except:
        filter_list.append(brand.lower())
    return df[df.Brand.isin(filter_list)]

#returns a subset of the dataframe filtered on any brand(s) needed AND any specific columns of interest
#e.g. Dataframe containing only Asda and Waitrose rows AND just the number of comments
def brandcolumns_mask(brand,cols='Brand'):
    X = brand_mask(brand)
    columns = [i.strip() for i in cols.split(',')]
    return X[columns]

#function that takes in a time period and brands as arguments and returns a summary dataframe of the number of posts any given brand
#made in the time frame specified
def time_frame_post(period='Year',brands='tesco,asda,waitrose,lidl,marks and spencer,sainsburys,morrisons'):
    brand_list = [i.lower() for i in brands.split(',')] 
    df_list = []
    for i in brand_list:
        x = pd.DataFrame(brand_mask(i).groupby([period])['Brand'].count())
        df_list.append(x)
    df_merged = reduce(lambda  left,right: pd.merge(left,right,on=[period],
                                            how='outer'), df_list)
    df_merged.columns = brand_list
    return df_merged.fillna(0).sort_index()

#string to use when I want to consider all the data and not look at individual brands
all_brands = 'sainsburys,lidl,tesco,morrisons,marks and spencer,waitrose,asda'
```

## `Exploratory Question` : Are social media metrics correlated in any way?
---

#### `Findings:`  

There seems to be a reasonably strong positive correlation (0.84) between responses to content (e.g. likes, haha, wow, love) and the number of shares the content received. Behaviourally this indicates that if people click on a response (with a haha for example) they are fairly likely to share it as well, at least more so than 'commenting-then-sharing' which has a weaker correlation co-efficient of 0.36.


```python
#function that takes in list of brands, x and y variables, ranges for x and y variables and plots a scatter plot for all brands 
#in their relevant brand colours
def scatter(brands,metric_x,metric_y,metric_x_max=10000,metric_y_max=10000,alpha=0.5):
    brand_frame = brand_mask(brands)
    plot_frame = brand_frame[(brand_frame[metric_x] <= metric_x_max) & ((brand_frame[metric_y] <= metric_y_max))]
    fig = plt.figure(figsize=(9, 9)) 
    ax = fig.gca()
    return sns.scatterplot(x=plot_frame[metric_x],y=plot_frame[metric_y],
                           hue=plot_frame.Brand,alpha=0.5,palette=brand_palette_list(brands))
```

```python
#plotting correlations between the three engagement metrics
fig, ax = plt.subplots(figsize=(4, 4))
sns.heatmap(brandcolumns_mask(all_brands,'Comments, Shares, All_Responses').corr(),annot=True,)
ax.set_xlabel('Engagement Metric')
ax.set_title('Relationship between engagement metrics - All brands')
plt.show()
```

Waitrose, Sainsburys and Morrisons are particulalry likely to post videos with at least half of their posts containing a video of some kind. Lidl and ASDA rarely posts videos.

```python
brand_summary_mean = df.groupby('Brand').mean()
brand_summary_count = df.groupby('Brand').sum()

#function that takes in a dataframe and a metric of choice and returns the frame ranked on that metric
def rank_frame(df,metric):
    return pd.DataFrame(df.sort_values(by=metric,ascending=False))

#function that takes in a dataframe and a metric of choice and plots how all the brands perform on that metric
def brand_barplot(df,metric):
    ranked = rank_frame(df,metric)    
    fig, ax = plt.subplots(figsize=(16, 4))
    title = '{} by Brand'.format(metric)
    colour_order = list(ranked.index)
    ax.set_xlabel('Brand')
    ax.set_ylabel(metric)
    ax.set_title(title)
    sns.barplot(x=ranked.index,y=ranked[metric],palette=brand_palette_list(colour_order))
    plt.show()
    return 
```

### `Exploratory Question: `  What are those hashtags saying?
---

#### `Findings:` 

The hashtags are usually hashtags relating to specific campaigns e.g #tasteofsummer (for Waitrose) or brand phrases e.g. #LidlSurprises (Lidl).

Morrisons use the hashtag #nutmeg a lot - this is their fashion line of clothing.

```python
from collections import Counter
#function that extracts any hashtag from a string using regular expressions and returns a dataframe
def hashtag_miner(x):
    hash_dic = dict(Counter(re.findall(r"#(\w+)",' '.join(list(x)))))
    tags = list(hash_dic.keys())
    count = list(hash_dic.values())
    return pd.DataFrame({'Hashtag' : tags,'Frequency' : count})
```

```python
#function takes in data frame, a brand and returns a plot with the brands colours
def hashtag_plot(df,brand):
    ranked = rank_frame(df,'Frequency').head(10)
    fig, ax = plt.subplots(figsize=(16, 2))
    title = '{} top 10 most frequently occuring hashtags'.format(brand)
    ax.set_xlabel('Hashtags')
    ax.set_ylabel('Frequency')
    ax.set_title(title)
    sns.barplot(x=ranked.Hashtag,y=ranked.Frequency,palette=brand_palette_list(brand),orient='v')
    sns.despine(bottom=True)
    plt.tight_layout(h_pad=2)
    plt.ylim(0, 100)
    return 
```

---

To calculate engagement the best way is to divide the number of any interactions by the reach of that post i.e.

A post that reached 10000 people and got 100 interactions we would say achieved an engagement rate of 0.01 (1%). 

However, reach data for posts are only viewable by admninistrators of any page. A way around this is to divide the interactions attained by the post by the total number of page likes a page has at the time of posting. *BIG CAVEAT* - This isn't a perfect solution as reach is not perfectly correlated with total page likes. But as long we are applying the same method to all brands then I can live with this.

## `Exploratory Question:` Are posts with videos more engaging than posts without videos?
---

#### `Findings:` 

The strip plot below shows that for most brands, any post that contains a video generally gets more shares (indicated by taller 'flames' on the right half of the chart). However for ASDA the story is the other way round i.e. Asda posts that contain no videos tend to get shared more. This is counter intuitive - generally posts with richer content get shared more (videos). This tells us something about the type of content ASDA is posting. We see a similar story for Lidl.

```python
#function that takes brands and metrics as arguments and returns a strip plot
def stripplot(brands,xmetric,metric_y,metric_y_max=1000000,alpha=0.4):
    brand_frame = brand_mask(brands)
    plot_frame = brand_frame[(brand_frame[metric_y] <= metric_y_max)]
    fig = plt.figure(figsize=(15, 9)) 
    ax = fig.gca()
    return sns.stripplot(x=plot_frame[xmetric],y=plot_frame[metric_y],
                           hue=plot_frame['Brand'],alpha=alpha,palette=brand_palette_list(brands),jitter=True,dodge=True,size=5)


stripplot(all_brands,'Contains_Video','Shares',10000)
```

## `Exploratory Question: `  When do brands start talking about Christmas?
---

#### `Findings:` 

In 2017, mentions of Christmas began in November and peaked in December. However in 2018, mentions came as early as October, peaked in November and dialled back a touch in December.

```python
#removes all punctuation
def remove_punc(x):\
    return re.sub(r'[^\w\s]','',x)
```

```python
#takes in a string of words (target) we want to count and a source and returns how many of the target 
#words occur in our source
def word_count(words, source):
    source_list = remove_punc(source).lower().split()
    word_list = [i.lower().strip() for i in words.split(',')]
    count = 0
    for i in word_list:
        if i in source_list:
            count +=1
    return count
```

```python
#apply the function to our Post Data and create a new column that returns how many times Christmas was mentioned in each post
df['Christmas'] = df.Post_Content.apply(lambda x: word_count('Christmas, christmastes, Xmas, festive, yule,Santa',x))
```

```python
#using groupby to look at how many Christmas counts we get by month for the past two years
xmas_2017 = df[(df.Year == 2017) ].groupby('Month')['Christmas'].sum()
xmas_2018 = df[(df.Year == 2018) ].groupby('Month')['Christmas'].sum()
df_xmas = pd.concat([xmas_2017, xmas_2018], axis=1)
df_xmas.columns = ['Christmas mentions 2017','Christmas mentions 2018']
```

```python
fig = plt.figure(figsize=(15, 6)) 
ax = fig.gca()
sns.lineplot(data=df_xmas)
ax.set_xlabel('Calendar Month')
ax.set_ylabel('Mentions of Christmas')
ax.set_title('Mentions of Christmas by month by year')
```
