```python
import pandas as pd
```

```python
morrisons_df = pd.read_csv('Morrisons_desktop.csv')
sainsburys_df = pd.read_csv('Sainsburys_desktop.csv')
tesco_df = pd.read_csv('Tesco_desktop.csv')
#asda_df = pd.read_csv('Asda_desktop.csv')
lidl_df = pd.read_csv('lidluk_desktop.csv')
mns_df = pd.read_csv('MarksandSpencer_desktop.csv')
waitrose_df = pd.read_csv('waitroseandpartners_desktop.csv')
```

```python
df = pd.concat([morrisons_df,sainsburys_df,tesco_df,lidl_df,mns_df,waitrose_df],axis=0)
```

```python
df
```

```python
df.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace= True)
df.drop_duplicates(inplace= True)
```
